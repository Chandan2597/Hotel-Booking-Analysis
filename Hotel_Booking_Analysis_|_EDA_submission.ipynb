## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

````python
# Visualization code

# Creating a function to get the count from a column
def get_count_from_column(df, column):
    new_df = df[column].value_counts()
    new_df = pd.DataFrame({'index':new_df.index, 'count':new_df.values})
    return new_df

# Creating a function to plot bar chart
def plot_bar_chart_from_column(df, column, title):
    new_df = get_count_from_column(df, column)
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(new_df['index'], new_df['count'], width=0.4, align='center', edgecolor='black', linewidth=2, color = ['g','r','b','c','y'])
    ax.set_xlabel(column)
    ax.set_ylabel('Count')
    ax.set_title(title)
    plt.xticks(rotation=90)
    plt.show()

# Creating a function to plot pie chart from a column
def plot_pie_chart_from_column(df, column, title, exp):
    new_df = get_count_from_column(df, column)
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.pie(new_df.loc[:,'count'], labels = new_df.loc[:,'index'], autopct='%1.2f%%', startangle=90, explode=exp, labeldistance=1)
    ax.set_title(title)
    plt.legend()
    ax.axis('equal')
    plt.show()
````
